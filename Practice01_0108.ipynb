```python
# https://tabelog.com/tw/tokyo/rstLst/?SrtT=rt
# https://tabelog.com/tw/tokyo/rstLst/1/?SrtT=rt
import urllib.request as req
# beautifulsoup4
import bs4 as bs

url = "https://tabelog.com/tw/tokyo/rstLst/1/?SrtT=rt"
response = req.urlopen(url)
# content = response.read()
# json.loads做型態轉換
html = bs.BeautifulSoup(response)
# print(html)
# 區塊型態: 兩個專屬功能
# find 找第一個 find_all 找所有
# 參數1: html名字 參數2: 篩選條件
link = html.find("a", {"class":"list-rst__rst-name-target"})
print(link)
# 找到你想要的區塊以後，你會想要萃取
# 特殊屬性(字典): link["href"]
# 2. 內容(專屬值): link.text
# 程式語言基本概念: 一個型態可以擁有兩種東西
# a. 專屬功能: 型態.功能() -> 人.吃飯(1hour)
# b. 專屬值: 型態.值 -> 人.姓名
print(link["href"])
print(link.text)
```

```python
import urllib.request as req
import bs4 as bs

url = "https://tabelog.com/tw/tokyo/rstLst/1/?SrtT=rt"
response = req.urlopen(url)
html = bs.BeautifulSoup(response)

# find_all: 形態是個類似list的
r_list = html.find_all("div", {"class":"list-rst__body"})
for r in r_list:
    en_name = r.find("a", {"class":"list-rst__rst-name-target"})
    en_name_href = en_name["href"]
    en_name_text = en_name.text

    print(en_name_text)
    print(en_name_href)
    print("-" * 30)

# link_list = html.find_all("a", {"class":"list-rst__rst-name-target"})
# for link in link_list:
#     print(link["href"])
#     print(link.text)
#     print("-" * 30)
```

```python
import urllib.request as req
import bs4 as bs


url = "https://tabelog.com/tw/tokyo/rstLst/1/?SrtT=rt"
response = req.urlopen(url)
html = bs.BeautifulSoup(response)

# find_all: 形態是個類似list的
r_list = html.find_all("div", {"class":"list-rst__body"})
for r in r_list:
    res_area = r.find("div", {"class":"list-rst__area-genre"})
    res_description = r.find("p", {"class":"list-rst__pr-title"})
    res_evaluate = r.find("span", {"class":"list-rst__rating-val"})
    res_price = r.find_all("span", {"class":"c-rating-v3__val"})
    res_dayoff = r.find("span", {"class":"list-rst__holiday-text"})

    print(res_area.text.strip())
    if res_description == None:
        print("無")
    else:
        print(res_description.text.strip())
        print(res_evaluate.text)
        print("pmprice: " + res_price[0].text)
        print("amprice: " + res_price[1].text)
        print("dayoff: " + res_dayoff.text)
        print("-" * 30)

# link_list = html.find_all("a", {"class":"list-rst__rst-name-target"})
# for link in link_list:
#     print(link["href"])
#     print(link.text)
#     print("-" * 30)
```
